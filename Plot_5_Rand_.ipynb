```python
import os
import glob
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
from collections import defaultdict

# ==========================================
# 1. SETUP & CONNECTION
# ==========================================
print("üîó Connecting to Google Drive...")
drive.mount('/content/drive')

BASE_PATH = '/content/drive/MyDrive/Colab Notebooks/CL-Drive'

if not os.path.exists(BASE_PATH):
    raise FileNotFoundError(f"‚ùå Path not found: {BASE_PATH}")

print(f"‚úÖ Data root found at: {BASE_PATH}")

# Define folder names
MODALITY_FOLDERS = {
    'eeg': 'EEG',
    'ecg': 'ECG',
    'eda': 'EDA',
    'gaze': 'Gaze',
    'labels': 'Labels'
}

# ==========================================
# 2. SMART SCANNING (FLEXIBLE)
# ==========================================
print("üîç Scanning directory structure flexibly...")

# Dictionary to store lists of participants found in each modality
participants_per_mod = defaultdict(list)
all_files_map = {} # To store file paths directly

for key, folder_name in MODALITY_FOLDERS.items():
    full_path = os.path.join(BASE_PATH, folder_name)
    if os.path.exists(full_path):
        # List all subdirectories (assuming structure: Modality/ParticipantID/files.csv)
        subdirs = [f for f in os.listdir(full_path) if os.path.isdir(os.path.join(full_path, f))]

        if not subdirs:
            # Maybe files are directly inside? Check files directly
            files_direct = glob.glob(os.path.join(full_path, "*.csv"))
            if files_direct:
                print(f"‚ö†Ô∏è No subfolders in {folder_name}, found {len(files_direct)} files directly.")
                # Treat as single participant or skip for now depending on logic
                participants_per_mod[key].append("ROOT")
            else:
                print(f"‚ö†Ô∏è Folder {folder_name} is empty or missing.")
        else:
            participants_per_mod[key] = subdirs
            print(f"   üìÇ {folder_name}: Found {len(subdirs)} participants.")
    else:
        print(f"   ‚ùå Folder {folder_name} NOT FOUND.")

# Find intersection (participants present in ALL found folders)
# If a folder is missing entirely, we ignore it for the intersection logic
available_modalities = [k for k, v in participants_per_mod.items() if v]

if not available_modalities:
    raise RuntimeError("‚ùå No data found in any modality folder!")

# Start with the first available modality's participants
common_participants = set(participants_per_mod[available_modalities[0]])

# Intersect with others
for mod in available_modalities[1:]:
    common_participants = common_participants.intersection(set(participants_per_mod[mod]))

common_participants = list(common_participants)

print(f"\nüìä Analysis:")
print(f"   Available modalities: {available_modalities}")
print(f"   Common participants across ALL available modalities: {len(common_participants)}")

if len(common_participants) == 0:
    print("\n‚ö†Ô∏è WARNING: No participant ID matches across all folders.")
    print("   Possible reasons:")
    print("   1. Folder names differ (e.g., '1030' vs 'participant_1030').")
    print("   2. Some modalities are missing data for these users.")
    print("   3. Structure is different.")

    # Fallback: Just pick random participants from the MOST populated modality (likely ECG)
    fallback_mod = max(available_modalities, key=lambda k: len(participants_per_mod[k]))
    print(f"   ‚û°Ô∏è Falling back to using participants from '{fallback_mod}' only.")
    selected_participants = random.sample(participants_per_mod[fallback_mod], min(5, len(participants_per_mod[fallback_mod])))
    force_single_mod = True
else:
    if len(common_participants) < 5:
        selected_participants = common_participants
    else:
        selected_participants = random.sample(common_participants, 5)
    force_single_mod = False

print(f"üé≤ Selected participants for visualization: {selected_participants}")

# ==========================================
# 3. HELPER FUNCTIONS
# ==========================================

def clean_csv(file_path):
    try:
        df = pd.read_csv(file_path, header=0)
        df = df.replace(r'^\s*$', np.nan, regex=True).fillna(method='ffill').fillna(0)
        return df.select_dtypes(include=[np.number])
    except Exception as e:
        return None

def find_file_flexible(p_id, level, mod_key, base_path):
    """Tries multiple strategies to find a file."""
    folder_name = MODALITY_FOLDERS[mod_key]

    # Strategy 1: Standard structure (Modality/Participant/Level)
    path1 = os.path.join(base_path, folder_name, p_id)
    if os.path.exists(path1):
        pattern = os.path.join(path1, f"*level_{level}*.csv")
        matches = glob.glob(pattern)
        if matches:
            # Prefer 'data' over 'baseline' if both exist
            data_matches = [m for m in matches if 'data' in m.lower()]
            return data_matches[0] if data_matches else matches[0]

    # Strategy 2: Maybe participant folder name is different?
    # (Skip for now to keep it simple, assuming IDs match if found in scan)

    return None

# ==========================================
# 4. PLOTTING LOGIC
# ==========================================

fig, axes = plt.subplots(nrows=len(selected_participants), ncols=5, figsize=(20, 4 * len(selected_participants)))
if len(selected_participants) == 1:
    axes = np.array([axes])

valid_rows = 0

for row_idx, p_id in enumerate(selected_participants):
    # Try to find a valid level randomly
    found_level = None
    files_map = {}

    levels_to_try = list(range(1, 10))
    random.shuffle(levels_to_try)

    for level in levels_to_try:
        temp_files = {}
        count_found = 0

        for mod_key in MODALITY_FOLDERS.keys():
            # If we are in fallback mode and this modality wasn't the main one, skip strict check
            if force_single_mod and mod_key != max(available_modalities, key=lambda k: len(participants_per_mod[k])):
                 # Still try to find, but don't fail if missing
                 pass

            f_path = find_file_flexible(p_id, level, mod_key, BASE_PATH)
            if f_path:
                temp_files[mod_key] = f_path
                count_found += 1

        # We need at least ECG + 1 other OR just ECG if fallback
        min_required = 1 if force_single_mod else 2

        if count_found >= min_required:
            # Ensure ECG is present if possible
            if 'ecg' in temp_files or force_single_mod:
                found_level = level
                files_map = temp_files
                break

    if found_level is None:
        print(f"‚ö†Ô∏è Skipping Participant {p_id}: No common level found.")
        for col_idx in range(5):
            axes[row_idx, col_idx].text(0.5, 0.5, f'No Data Found\nFor P-ID: {p_id}', ha='center', va='center', color='red')
            axes[row_idx, col_idx].axis('off')
        continue

    print(f"‚úÖ Plotting Participant: {p_id} | Level: {found_level} | Files found: {list(files_map.keys())}")
    valid_rows += 1

    plot_order = ['eeg', 'ecg', 'eda', 'gaze', 'labels']
    titles = ['EEG', 'ECG', 'EDA', 'Gaze', 'Label']

    for col_idx, mod_key in enumerate(plot_order):
        ax = axes[row_idx, col_idx]
        file_path = files_map.get(mod_key)

        if file_path:
            df = clean_csv(file_path)
            if df is not None and not df.empty:
                if mod_key == 'labels':
                    y_vals = df.iloc[:, 0].values
                    x_vals = np.arange(len(y_vals))
                    ax.step(x_vals, y_vals, where='mid', color='black', linewidth=2)
                    ax.set_ylim(0, 10)
                    ax.set_yticks(range(1, 10))
                    ax.grid(True, alpha=0.3)
                else:
                    cols_to_plot = [c for c in df.columns if df[c].sum() != 0]
                    if cols_to_plot:
                        limit = min(4, len(cols_to_plot))
                        for i, col in enumerate(cols_to_plot[:limit]):
                            y_vals = df[col].values
                            x_vals = np.arange(len(y_vals))
                            offset = i * (np.max(y_vals) - np.min(y_vals)) * 1.2 if len(cols_to_plot) > 1 else 0
                            ax.plot(x_vals, y_vals + offset, label=str(col)[:8], linewidth=0.8, alpha=0.8)
                        ax.grid(True, alpha=0.3)
                        if len(cols_to_plot) > 1:
                            ax.legend(fontsize=6)
            else:
                ax.text(0.5, 0.5, 'Empty/Clean Failed', ha='center', va='center', color='gray')
                ax.axis('off')
        else:
            ax.text(0.5, 0.5, 'File Missing', ha='center', va='center', color='orange', style='italic')
            ax.axis('off')

        if row_idx == 0:
            ax.set_title(titles[col_idx], fontsize=12, fontweight='bold')
        if col_idx == 0:
            ax.set_ylabel(f'ID: {p_id}\nLvl: {found_level}', fontsize=9)
        ax.set_xticks([])

if valid_rows == 0:
    plt.suptitle("No valid data could be plotted!", fontsize=16, color='red')
else:
    plt.suptitle(f'Visualization of {valid_rows} Participants (Flexible Scan)', fontsize=16, fontweight='bold')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

print("\n‚ú® Done! If you see orange 'File Missing' boxes, it means that specific modality is not uploaded for these users yet.")
```
